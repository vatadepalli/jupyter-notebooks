```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

## Babies

```python
us_babies = pd.read_csv("data/linkedin/us_baby_names.csv")
us_babies
```

```python
us_babies_2014 = us_babies.loc[us_babies['Year'] == 2014, :]
sorted_us_2014 = us_babies_2014.sort_values('Count', ascending = False)
sorted_us_2014.iloc[0: 5]
```

```python
states_babies = pd.read_csv('data/linkedin/state_baby_names.csv')
states_babies
```

```python
states_babies_2014 = states_babies.loc[states_babies['Year'] == 2014, :]
ca_babies_2014 = states_babies_2014.loc[states_babies_2014['State'] == 'CA', : ]
ca_babies_2014.sort_values('Count', ascending = False)
ca_babies_2014.iloc[0: 5]
```

```python
def popular(series):
    """Recieves series of baby names with descending order on count.
    Returns the most popular name in series"""
    return series.iloc[0]

ca_babies = states_babies.loc[states_babies['State'] == 'CA', : ]
ca_grouped = ca_babies.sort_values('Count', ascending = False).groupby(['Year', 'Gender']).agg(popular)
ca_grouped
```

```python
us_aditya = us_babies.loc[us_babies['Name'] == 'Aditya', : ]
us_aditya.plot.barh(x = "Year", y = "Count")
```

## Airbnb

```python
listings = pd.read_csv('data/linkedin/Airbnb_NYC_2019.csv')
listings
```

```python
sn.countplot(x = 'neighbourhood_group', data = listings)
plt.show
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings, ci = False)
plt.show()
```

```python
plt.hist(listings['price'])
plt.xlabel('price (in US dollars)')
plt.show()
```

```python
plt.hist(listings['price'], bins = np.arange(0, 1100, 40))
plt.xlabel('price (in US dollars)')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0, 1100)
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'], s = 5)
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0, 1100)
plt.show()
```
