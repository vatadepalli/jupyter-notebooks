# PyPlot

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['figure.dpi'] = 600
data = pd.read_csv('./data/baseballdb/core/Batting.csv')

df = data[['HR', 'RBI', 'BB', 'playerID',  'yearID', 'teamID']]
df = df[df['yearID'] >= 2000].sample(10).sort_values('yearID')

df.plot(x = 'yearID', y = 'HR')

plt.title('HR vs. Year')
plt.xlabel('Year')
plt.ylabel('HR Count')

plt.savefig('df.png')  # save to PNG file
plt.show()

display(df)
```

```python
df.set_index('yearID', inplace = True)
display(df)
df.plot()
df.plot(kind='box')
plt.show()
```

```python
df['HR'].plot(kind='hist', x = 'HR')
plt.show()

df.plot(kind='box', x = 'HR')
plt.show()
```
